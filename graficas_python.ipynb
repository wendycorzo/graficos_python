# GRAFICAS CON LA LIBRERIA MATPLOTLIB

## GRAFICAS EN DOS DIMENSIONES

### Gráfica de una lista

```python
# importamos la librería
# import matplotlib.pyplot

from matplotlib.pyplot import plot, show,xlim, ylim, xticks
```

```python
# Grafica N°1

y = [1,2,3,4,5]

plot(y)
show()
```

```python
# Grafica N°2
y = [1,2,3,4,5]

plot(y, '*')
show()
```

```python
# Grafica N°3

x = [1,2,3,4,5]
y = [2,4,6,8,10]

plot(x, y, '*')
show()
```

```python
# Grafica N°4

# Grafica con límites

x = [1,2,3,4,5]
y = [2,4,6,8,10]

plot(x, y, '*')
xlim(-10,10)
ylim(0, 15)
show()
```

```python
 #Grafica N°5

from numpy import linspace, pi, sin

x = linspace(-10*pi, 10*pi, 2500)
y = sin(x)

plot(x,y)
ylim(-2,2)
show()
```

```python
 #Grafica N°6

from numpy import linspace, pi, sin

x = linspace(-pi, pi, 256)
y = sin(x)

plot(x,y)
ylim(-2,2)
xticks([-pi,-pi/2,0,pi/2,pi],[r'$-\pi$',r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
show()
```
